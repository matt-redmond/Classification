```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
#questions about importing from sklearn
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
```

```python
df = pd.read_csv('mushrooms.csv')

df["class_num"] = df["class"].map(lambda x: 1 if (x == "p") else 0)

X = df.loc[:,'cap-shape':'habitat']
y = df['class_num']

X = pd.get_dummies(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state=42)

lr_model = LogisticRegression(C=1)
lr_model.fit(X_train, y_train)

tc = confusion_matrix(y_test, lr_model.predict(X_test))
plt.figure(dpi=150)
sns.heatmap(tc, cmap=plt.cm.Blues, annot=True, square=True)

print(lr_model.score(X_train, y_train))
print(lr_model.score(X_test, y_test))
```
